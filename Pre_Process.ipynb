### This notebook shows that one has to experiment with the image to create the best epicycle animation

```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
import cv2
```

### TIGER

```python
img = cv2.imread('./Images/TIGER.png',0)
img = cv2.resize(img,(500,500))
img.shape
```

```python
plt.imshow(img,'gray')
```

```python
ret,thresh1 = cv2.threshold(img,160,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)

titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]


plt.figure(figsize=(10,5))

for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```

#### Gettin the best animation depends on the code below and different combination for different images
##### Line arts, make the best animation

```python
img = cv2.imread('./Images/TIGER.png',0)
img = cv2.resize(img,(200,200))


_,edges = cv2.threshold(img,100,255,cv2.THRESH_TRUNC)
#edges = cv2.GaussianBlur(edges, (5,5), 0)
#edges = cv2.erode(edges, None, iterations=1)
#edges = cv2.dilate(img, None, iterations=5)
edges = cv2.Canny(img,10,80)
plt.imshow(edges,'gray')
```

#### Once satisfied , paste the above snippet to main.py

```python
contours, hier = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
img = cv2.imread('./Images/ROSE.jpg',0)
plt.imshow(img,'gray')
```

### HORSE

```python
img = cv2.imread('./Images/HORSE.png',0)
plt.imshow(img,'gray')
```

```python
img = cv2.imread('./Images/HORSE.png',0)
img = cv2.resize(img,(300,200))
_,edges = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)

edges = cv2.GaussianBlur(edges, (5,5), 0)
#edges = cv2.erode(edges, None, iterations=1)
edges = cv2.dilate(edges, None, iterations=1)
edges = cv2.Canny(edges,10,120)
plt.imshow(edges,'gray')
```

```python
contours, hier = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```
